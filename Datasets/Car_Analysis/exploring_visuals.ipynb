# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# load datasets

df_08 = pd.read_csv('clean_08.csv')
df_18 =pd.read_csv('clean_18.csv')
```

```python
df_08.columns
```

```python
df_08.greenhouse_gas_score.hist(figsize = (8,8), label = '2008', alpha = 0.5)
df_18.greenhouse_gas_score.hist(figsize = (8,8), label = '2018', alpha = 0.5)
plt.title('Greenhouse Gas Score')
plt.legend();
```

```python
df_08.cmb_mpg.hist(figsize = (8,8), label = '2008', alpha = 0.5)
df_18.cmb_mpg.hist(figsize = (8,8), label = '2018', alpha = 0.5)
plt.legend()
plt.title('Combined Miles per Gallons');
```

```python
fig = plt.figure(figsize = (10,8))

axl = fig.add_subplot(111)
axl.scatter(x = df_08['displ'], y = df_08['cmb_mpg'], c = 'b', label = '2008', alpha = 0.5)
axl.scatter(x = df_18['displ'], y = df_18['cmb_mpg'], c = 'r', label = '2018', alpha = 0.5)
plt.legend(loc = 'upper right')
plt.xlabel('Displacement')
plt.ylabel('Combined MPG')
plt.title('Correlation Between Combined MPG and Displacement')
plt.show()
```

```python
fig2 = plt.figure(figsize = (10,8))

axl = fig2.add_subplot(111)
axl.scatter(x = df_08['greenhouse_gas_score'], y = df_08['cmb_mpg'], c = 'b', label = '2008', alpha = 0.5)
axl.scatter(x = df_18['greenhouse_gas_score'], y = df_18['cmb_mpg'], c = 'r', label = '2018', alpha = 0.5)
plt.legend(loc = 'upper right')
plt.ylabel('Combined MPG')
plt.xlabel('Greenhouse Gas Score')
plt.title('Correlation Between Greenhouse Gas Score and Combined MPG')
plt.show()
```
