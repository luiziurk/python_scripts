**Task 2**: Plot the relationship between engine size ('displ', in liters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
# YOUR CODE HERE
fuel_econ.fuelType.value_counts()
```

```python
#The code below allows to filter only on fuel types which count is above a determined value
fuel_econ_filtered = fuel_econ.groupby("fuelType").filter(lambda x: len(x)>1000)
```

```python
fuel_econ_filtered.fuelType.value_counts()
```

> Perfoming the filter brings only the desired fuel types

```python
plt.Figure(figsize=(12,10))
sb.boxplot(data = fuel_econ_filtered,  x = 'VClass', y = 'displ', hue = 'fuelType')
plt.xticks(rotation = 15)
plt.ylabel("Engine Size (l)")
plt.xlabel("Vehicle Type")
plt.legend(loc = 6, bbox_to_anchor = (1.0, 0.5));
```
