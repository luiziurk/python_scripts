```python
df_fuel_econ_filt = fuel_econ.groupby('make').filter(lambda x : len(x) >= 80)
df_fuel_econ_filt.reset_index(drop = True)
df_fuel_econ_filt.head()
```

```python
# Will plot ordering by the most frequent brand to the least frequent
g = sb.FacetGrid(data=df_fuel_econ_filt, col='make',col_wrap=4, col_order=df_fuel_econ_filt["make"].value_counts().index)
g.map(plt.hist, 'comb')
g.fig.suptitle("Combined Fuel Consuption (mpg) by Brand (Most Frequent to Least)", y = 1.03);
```

Continuing on from the previous task, plot the mean fuel efficiency for each manufacturer with at least 80 cars in the dataset.

```python
df_means = df_fuel_econ_filt.groupby('make').mean()['comb'].sort_values(ascending=False)
df_means.index
```

```python
fig = plt.figure(figsize=(12,10))

plt.title("Average of Combined Fuel Consumption (mpg) by Brand")
sb.barplot(data = df_fuel_econ_filt, x = 'comb', y = 'make', color=sb.color_palette()[0],order=df_means.index,errorbar='sd')
plt.xlabel("Average Fuel Consumption (mpg)")
plt.ylabel("Brand");
```
